```python
# This code may not be run on GFG IDE
# as required packages are not found.
	
# STRATIFIES K-FOLD CROSS VALIDATION { 10-fold }

# Import Required Modules.
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold

# create StratifiedKFold object
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
lst_accu_stratified = []

# store all splits
X_train = []
X_test = []
y_train = []
y_test = []

for train_index, test_index in skf.split(test_model.X_scaled, test_model.y):
	# split X and y
	x_train_fold, x_test_fold = test_model.X_scaled[train_index], test_model.X_scaled[test_index]
	y_train_fold, y_test_fold = test_model.y[train_index], test_model.y[test_index]
	
	# store splits
	X_train.append(x_train_fold)
	X_test.append(x_test_fold)
	y_train.append(y_train_fold)
	y_test.append(y_test_fold)
	
	# 
	test_model.user_defined_model.fit(x_train_fold, y_train_fold)
	lst_accu_stratified.append(test_model.user_defined_model.score(x_test_fold, y_test_fold))

# Print the output.
print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
	max(lst_accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
	min(lst_accu_stratified)*100, '%')
print('\nOverall Accuracy:',
	mean(lst_accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```
